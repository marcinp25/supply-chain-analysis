```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries loaded ✅")
```

```python
df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding="latin1")

print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
df.head()
```

```python
print("=== Data Types ===")
print(df.dtypes)
print()
print("=== Missing Values ===")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Distribution of key categorical columns
print("=== Delivery Status ===")
print(df["Delivery Status"].value_counts())
print()
print("=== Shipping Mode ===")
print(df["Shipping Mode"].value_counts())
print()
print("=== Market ===")
print(df["Market"].value_counts())
```

```python
# Key numeric columns summary
cols = ["Sales", "Order Profit Per Order", "Days for shipping (real)",
        "Days for shipment (scheduled)"]
print(df[cols].describe().round(2))
```

```python
# Distribution of delay days
df["delay_days_temp"] = df["Days for shipping (real)"] - df["Days for shipment (scheduled)"]

df["delay_days_temp"].value_counts().sort_index().plot(
    kind="bar",
    figsize=(8, 4),
    color="#2E75B6",
    edgecolor="white"
)
plt.title("Distribution of Delay Days")
plt.xlabel("Delay Days")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

df = df.drop(columns=["delay_days_temp"])
```

```python
# Removing unusable columns
df = df.drop(columns=["Order Zipcode", "Product Description", "Customer Password", "Product Image"])

print(f"Columns after cleaning: {df.shape[1]}")
```

```python
# Converting dates to datetime
df["order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"])
df["shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"])

print("Date types after conversion:")
print(df[["order date (DateOrders)", "shipping date (DateOrders)"]].dtypes)
```

```python
# Feature engineering - delivery delay in days
df["delay_days"] = df["Days for shipping (real)"] - df["Days for shipment (scheduled)"]

print("delay_days summary:")
print(df["delay_days"].describe())
```

```python
df.to_csv("dataco_cleaned.csv", index=False)

print(f"Cleaned dataset saved: {df.shape[0]} rows, {df.shape[1]} columns")
```
